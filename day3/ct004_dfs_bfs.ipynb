#### 재귀호출
- 꽤 어렵다. 잘 생각해보면 그렇게 어렵지도 않음
- 디버깅이 힘들기 때문에 어렵다고 생각됨
- 재귀호출을 사용하지 않으면 소스코드가 아주 많이 길어질 수 있음
- 수학적인 지식이 많이 들어가는 함수

```python
## maximum recursion depth exceeded while calling a Python object
## 어느 이상이 지나면 예외발생. 무한대로 재귀호출을 할 수 없음
## 다른언어는 컴퓨터 메모리가 받쳐줄때까지 무한대로 재귀!
## 프랙탈 구조(시에르핀스키 삼각형 등) 그리거나 할 때 재귀호출 사용

def rec_call():
    print('재귀호출')
    rec_call()

rec_call()
```

```python
## 일반적인 방식(반복)으로 팩토리얼 구하기
def factorial_iter(n):
    result = 1
    # 1부터 n까지의 수를 곱하기
    for i in range(1, n+1):
        result *= i
    return result ## 팩토리얼이기 때문에 짧음.
```

```python
## 재귀호출로 팩토리얼 구하기
def factorial_rec(n):
    if n <= 1:
        return 1
    
    # n! = n(n-1)!를 구하기. 수학에서 쓰는 공식 그대로 쓰는 방식
    return n * factorial_rec(n - 1)
```

```python
## 430자리 이상의 수는 못 구함
factorial_iter(1550)
```

```python
## C#, Java에서는 실행 시간이 오래 걸림
factorial_rec(1550)
```

#### DFS
- Depth First Search : 깊이 우선 탐색
    - 그래프에서 깊은 부분을 우선적으로 탐색하는 알고리즘

- 노드와 엣지 간의 관계를 표현하는 방식
    - 인접행렬
    - 인접리스트

```python
## 인접행렬(매트릭스 형태)
INF = 999999999 # 무한대 뜻

graph = [ 
    [0, 7, 5],
    [7, 0, INF],
    [5, INF, 0]
]

graph
```

- 무방향 그래프에서는 사선으로 0값이 연결됨

```python
## 인접리스트(리스트 형태)
graph2 = [[] for _ in range(3)] # 노드가 3개니까

## 노드 0에서 가는 길을 입력
graph2[0].append((1, 7)) ## 1번 노드로 갈 때 비용 7
graph2[0].append((2, 5)) ## 2번 노드로 갈 때 비용 5

## 노드 1번에서 갈 수 있는 길 입력
graph2[1].append((0,7))

## 노드 2번에서 갈 수 있는 길
graph2[2].append((0, 5))

graph2
```

- 인접행렬
    - 무방향일때 효율적
    - 메모리를 많이 씀
- 인접리스트
    - 방향성이 있을 때 효율적
    - 메모리 효율적

테스트 코드 예제그래프

```python
## 인접리스트로 그래프값을 구성
dfs_graph = [
    [], ## 0번 노드는 없기 때문에 비워둠
    [2, 3, 8], # 1번 노드에서는 2, 3, 8번 노드로 갈 수 있음
    [1, 7],
    [1, 4, 5],
    [3, 5],
    [3, 4],
    [7],
    [2, 6, 8],
    [1, 7]
]

## 각 노드에 방문정보를 1차원 리스트로 생성
visited = [False] * 9

dfs_graph
```

```python
## DFS 함수
def dfs(graph, node, visited):
    # 현재 노드를 방문처리
    visited[node] = True
    print(node, end=' ') # 재귀함수이기 때문에 마지막 호출된 재귀함수의 값이 출력
    # 현재 노드와 연결된 다른 노드를 재귀적으로 방문(호출)
    for i in graph[node]:
        if not visited[i]:
            ## 재귀호출!!!
            dfs(graph=graph, node=i, visited=visited)
```

```python
dfs(dfs_graph, 1, visited)
```

#### BFS
- Breadth First Search : 넓이 우선 탐색
    - 가까운 노드부터 탐색하는 알고리즘
    - 큐를 사용해서 구현
    - $O(N)$ 복잡도, DFS보다 더 빠르다

- 재귀호출은 컴퓨터 수행시간이 빠르지 않음. 느려지는 경우도 발생

```python
bfs_graph = [
    [], ## 0번 노드는 없기 때문에 비워둠
    [2, 3, 8], # 1번 노드에서는 2, 3, 8번 노드로 갈 수 있음
    [1, 7],
    [1, 4, 5],
    [3, 5],
    [3, 4],
    [7],
    [2, 6, 8],
    [1, 7]
]

## 각 노드에 방문정보를 1차원 리스트로 생성
bfs_visited = [False] * 9

bfs_graph
```

```python
def bfs(graph, start, visited):
    ## 큐 대신 deque
    queue = deque([start])
    ## 현재 노드 방문처리
    visited[start] = True
    ## 큐가 빌 때까지 반복
    while queue:
        ## 큐에서 하나의 노드를 뽑는다
        node = queue.popleft()
        print(node, end=' ')
        ## 해당 노드가 연결된 아직 방문 안한 노드들을 큐에 삽입
        for i in graph[node]:
            if not visited[i]: ## 방문하지 않았으면
                queue.append(i)
                visited[i] = True
```

```python
bfs(bfs_graph, 1, bfs_visited)
```

- DFS(스택, 재귀호출) 보다 BFS(큐)가 쉽고 간단하다.
- 코딩테스트는 DFS가 더 많이 나옴
